```python
def get_location(division):
   try:
       return division.find('span', {'class':'location'}).text
   except:
       return 'NA'
   
def get_summary(division):
   try:
       return division.find('span', {'class':'summary'}).text
   except:
       return 'NA'

def get_jobtitle(division):
   try:
       return division.find('a', {'data-tn-element':'jobTitle'}).text
   except:
       return 'NA'
   
def get_sallary(division):
   try:
       return division.find('span', {'class':'no-wrap'}).text
   except:
       return 'NA'
   
def get_company(division):
   try:
       return division.find('span', {'class':'company'}).text
   except:
       return 'NA'
   
def get_date(division):
   try:
       return division.find('span', {'class':'date'}).text
   except:
       return 'NA'
job_data = pd.DataFrame(columns=['city', 'state', 'job_title', 'company', 'salary', 'job_description','job_type','salary_category', 'date'])
job_type=['parttime','fulltime','contract','internship','commission','temporary']
salary_category=['%2480%2C000','%24100%2C000','%24120%2C000','%24140%2C000','%24160%2C000']
job_list=[]

for x in range(0,100,10):
   for j in job_type:
       for salary in salary_category:
           url ="https://au.indeed.com/jobs?q=data+scientist%2C+data+analyst%2C+research+scientist%2C+business+intelligence%2C+machine+learning%2C+python+"+salary+"&l=Australia&jt="+j+"&start="+str(x)
           print(url)
           html = requests.get(url)
           soup = BeautifulSoup(html.text, 'html.parser')
           for sp1 in soup.find_all('div', {'class':' row result'}):
               job_list.append(sp1)
           time.sleep(1)
           
           for division in job_list:
               location=get_location(division)
               job_title = get_jobtitle(division)
               company_Name = get_company(division)
               salaries = get_sallary(division)
               summary = get_summary(division)
               date=get_date(division)
               job_type=j
               salary_category=salary
               job_data.loc[len(job_data)] = [location, 'Nan', job_title, company_Name, salaries, summary,job_type,salary_category, date]
           time.sleep(1)
job_data.to_csv('my_new_job.csv')
```
