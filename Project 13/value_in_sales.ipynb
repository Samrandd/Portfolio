### Creating Cost, Sales, Profit and Markup per transaction columns 

```python
data['CostPerTransaction']=data['NumberOfItemsPurchased']*data['CostPerItem']
```

```python
data['SalePerTransaction']=data['SellingPricePerItem']*data['NumberOfItemsPurchased']
```

```python
data['ProfitPerTransaction']=data['SalePerTransaction']-data['CostPerTransaction']
```

```python
data['Markup']=round(data['ProfitPerTransaction']/data['CostPerTransaction'],2)
```

```python
#create date field
#my_date='Day'+'-'+'Month'+'-'+'Year'
```

```python
data['month'] = pd.to_datetime(data.Month, format='%b').dt.month
```

```python
data['date']=data['Day'].astype(str)+'-'+data['month'].astype(str)+'-'+data['Year'].astype(str)
```

### Data Cleaning

```python
#Split Client Keyword
```

```python
split_col=data['ClientKeywords'].str.split(',',expand=True)
```

```python
data['ClientAge']=split_col[0]
```

```python
data['ClientType']=split_col[1]
```

```python
data['LengthofContract']=split_col[2]
```

```python
#remove bracket
data['ClientAge']=data['ClientAge'].str.replace('[','')
```

```python
#LengthofContract
data['LengthofContract']=data['LengthofContract'].str.replace(']','')
```

```python
data['ItemDescription']=data['ItemDescription'].str.lower()
```

```python
#Joinning two tables
seasons=pd.read_csv("value_inc_seasons.csv",sep=';')
```

```python
data=pd.merge(data,seasons,on='Month')
```

```python
data.head(5)
```

```python
#drop unused data
data=data.drop(['ClientKeywords','Day','Year','Month'],axis=1)
```
